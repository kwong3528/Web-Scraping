```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

# 상세페이지 url

```python
test_url = []
for i in range(1,3): # 사이트의 가져올 페이지의 수다. ex 1페이지부터 2페이지를 뜻한다.
    html = requests.get("사이트주소(사이트의 주소의 규칙성부분을 찾은뒤 그것을 i에 담아 순회시킨다).".format(i)).text
    soup = BeautifulSoup(html, "html.parser")
    book_list = soup.find_all("dt")[2:]

    for w in range(len(book_list)):
        url = book_list[w].find("a")["href"]
        test_url.append(url)
```

```python
dic = {"Title": [],"Context": [],"Category": [], "Company": []}

for i in range(len(test_url)):
    req = requests.get(test_url[i])
    html = req.text
    soup = BeautifulSoup(html, "html.parser")
    
    texts = soup.find_all("div", {"class": "recommDesc"})
    for i in texts:
        text_list = soup.find_all("p")[1]
        text_list = text_list.text
        text_list = re.sub("[\n\t\r]","",text_list)
        
    title = soup.find_all("strong")[0].text
    k = soup.find_all("ul",{"class","dot-list clearfix"})
    
    information = (k[0].find_all("li"))
    category = information[0].text
    company = information[2].text
    
    
    dic["Title"].append(title)
    dic["Category"].append(category)
    dic["Company"].append(company)
    dic["Context"].append(text_list)
```

# 카테고리와 출판사부분의 형식을 정규화하기

```python
clean_text = []
for i in dic["Category"]:
    result = re.sub("['분야 :']","",i.strip()) # 불필요한 부분이 내용에 있어 그부분을 전처리 하였다.
    clean_text.append(result)
dic["Category"] = clean_text

clean_text2 = []
for i in dic["Company"]:
    result = re.sub("['출판사 :']","",i.strip())
    clean_text2.append(result)
dic["Company"] = clean_text2
```
